```python
import tkinter as tk
from tkinter import messagebox

# Function to handle a player's choice
def make_choice(player, choice):
    global player1_choice, player2_choice

    if player == 1:
        player1_choice = choice
        player1_label.config(text=f"Player 1: {choice}")
    elif player == 2:
        player2_choice = choice
        player2_label.config(text=f"Player 2: {choice}")

    # Check if both players have made their choices
    if player1_choice and player2_choice:
        determine_winner()

# Function to determine the winner
def determine_winner():
    global player1_choice, player2_choice

    # Winning conditions
    rules = {"Rock": "Scissors", "Scissors": "Paper", "Paper": "Rock"}

    if player1_choice == player2_choice:
        result = "It's a tie!"
    elif rules[player1_choice] == player2_choice:
        result = "Player 1 wins!"
    else:
        result = "Player 2 wins!"

    messagebox.showinfo("Result", result)
    reset_game()

# Function to reset the game
def reset_game():
    global player1_choice, player2_choice
    player1_choice = None
    player2_choice = None
    player1_label.config(text="Player 1: ")
    player2_label.config(text="Player 2: ")

# Initialize the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors Game")

# Set window background color
root.configure(bg="lightblue")

# Initialize player choices
player1_choice = None
player2_choice = None

# Style configurations
label_font = ("Arial", 14, "bold")
button_font = ("Arial", 12, "bold")
player_label_bg = "lightyellow"
button_bg = {"Rock": "red", "Paper": "blue", "Scissors": "green"}
button_fg = "white"

# Player labels
player1_label = tk.Label(root, text="Player 1: ", font=label_font, bg=player_label_bg)
player1_label.grid(row=0, column=0, padx=10, pady=10)

player2_label = tk.Label(root, text="Player 2: ", font=label_font, bg=player_label_bg)
player2_label.grid(row=1, column=0, padx=10, pady=10)

# Buttons for Player 1
tk.Label(root, text="Player 1 Choices:", font=label_font, bg="lightblue", fg="black").grid(row=0, column=1, padx=10)
for i, choice in enumerate(["Rock", "Paper", "Scissors"]):
    button = tk.Button(root, text=choice, font=button_font, bg=button_bg[choice], fg=button_fg,
                       command=lambda c=choice: make_choice(1, c))
    button.grid(row=0, column=i+2, padx=5)

# Buttons for Player 2
tk.Label(root, text="Player 2 Choices:", font=label_font, bg="lightblue", fg="black").grid(row=1, column=1, padx=10)
for i, choice in enumerate(["Rock", "Paper", "Scissors"]):
    button = tk.Button(root, text=choice, font=button_font, bg=button_bg[choice], fg=button_fg,
                       command=lambda c=choice: make_choice(2, c))
    button.grid(row=1, column=i+2, padx=5)

# Run the application
root.mainloop()
```
